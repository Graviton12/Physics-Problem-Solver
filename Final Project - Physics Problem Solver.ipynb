```python
import math

#for Kinematic equations
class Kinematics():
    #initializing paramters/variables
    def __init__(self, initial_velocity=None, final_velocity=None, acceleration=None, time=None, displacement=None):
        self.initial_velocity = initial_velocity
        self.final_velocity = final_velocity
        self.acceleration = acceleration
        self.time = time
        self.displacement = displacement  
    
    #function that solves for the unknown paramter/variable
    def solve(self):
        #prints out the following derived kinematic equations for the user's reference
        print("\n--- Derived kinematic Equations ---")
        print("1. ΔX = (1/2)(Vi + Vf)t")
        print("2. ΔX = Vit + (1/2)(a)(t^2)")
        print("3. Vf = sqrt(Vi^2 + 2aΔX)")
        print("4. Vf = Vi + at")
        print("5. Vi = Vf - at")        
        print("6. a = 2(ΔX - Vit)/(t^2)")
        print("7. t = (Vf - Vi)/a")        
        
        #prompting the user to enter known variables and leave unknown parameters blank
        print("\nLeave the parameter you are looking for and any parameters not required blank. Enter the known variables:")
        self.initial_velocity = float(input("Enter the Initial velocity (m/s): ") or 0)
        self.final_velocity = float(input("Enter the Final velocity (m/s): ") or 0)
        self.acceleration = float(input("Enter the Acceleration (m/s^2): ") or 0)
        self.time = float(input("Enter the Time (s): ") or 0)
        self.displacement = float(input("Enter the Displacement (m): ") or 0)        
        
        #conditions depending on which variables were left blank and which were entered.
        #conditon to solve for equation 1. displacement
        if self.displacement == 0 and self.initial_velocity != 0 and self.final_velocity != 0 and self.time != 0:
            self.displacement_solve1()
        #condition to solve for equation 2. displacement
        elif self.displacement == 0 and self.initial_velocity != 0 and self.acceleration != 0 and self.time != 0:
            self.displacement_solve2()
        #condition to solve for equation 3. initial velocity
        elif self.initial_velocity == 0 and self.final_velocity != 0 and self.acceleration !=0 and self.time != 0:
            self.initial_velocity_solve()
        #condition to solve for equation 4. final velocity
        elif self.final_velocity == 0 and self.initial_velocity != 0 and self.acceleration !=0 and self.time != 0:
            self.final_velocity_solve1()
        #condition to solve for equation 5. final velocity
        elif self.final_velocity == 0 and self.initial_velocity != 0 and self.acceleration !=0 and self.displacement != 0:
            self.final_velocity_solve2()  
        #condition to solve for equation 6. acceleration
        elif self.acceleration == 0 and self.displacement != 0 and self.initial_velocity != 0 and self.time != 0:
            self.acceleration_solve()
        #condition to solve for equation 7. time
        elif self.time == 0 and self.final_velocity != 0 and self.initial_velocity != 0 and self.acceleration !=0:
            self.time_solve()
        #condition if unable to compute
        else:
            print("Unable to solve for unknown variable due to lack of information.") 
    
    #functions that calculate the unknown variable that the user is looking for
    def displacement_solve1(self):
        self.displacement = (self.initial_velocity + self.final_velocity)*(.5)*(self.time)
        print("The displacement is:", self.displacement, "m")
    def displacement_solve2(self):
        self.displacement = (self.initial_velocity * self.time)+(.5)*(self.acceleration)*(math.pow(self.time, 2))
        print("The displacement is:", self.displacement, "m")
    def initial_velocity_solve(self):
        self.initial_velocity = self.final_velocity - (self.acceleration*self.time)
        print("The initial velocity is: ", self.initial_velocity, "m/s")
    def final_velocity_solve1(self):
        self.final_velocity = self.initial_velocity + (self.acceleration*self.time)
        print("The final velocity is: ", self.final_velocity, "m/s")
    def final_velocity_solve2(self):
        self.final_velocity = math.sqrt(math.pow(self.initial_velocity, 2) + (self.acceleration * self.displacement))
        print("The final velocity is: ", self.final_velocity, "m/s")
    def acceleration_solve(self):
        self.acceleration = 2*((self.displacement) - (self.initial_velocity*self.time)) / (math.pow(self.time, 2))
        print("The acceleration is: ", self.acceleration, "m/s^2")
    def time_solve(self):
        self.time = (self.final_velocity - self.initial_velocity) / (self.acceleration)
        print("The time is: ", self.time, "s")

#for Projectile Motion equations
class Projectile_motion():
    #initializing paramters/variables
    def __init__(self, flight_time=None, max_height=None, horizontal_range=None, initial_velocity=None, theta=None, acceleration=None):
        self.flight_time=flight_time
        self.max_height=max_height
        self.horizontal_range=horizontal_range
        self.initial_velocity=initial_velocity
        self.theta=theta
        self.acceleration=acceleration
    
    #function that solves for the unknown paramter/variable
    def solve(self):
        #prints out the following projectile motion equations for the user's reference
        print("\n--- Projectile Motion Equations ---")
        print("1. t = 2Visin(θ)/g")
        print("2. H = vi^2sin^2(θ)/2g")
        print("3. R = vi^2sin(2θ)/g")
        
        #prompts the user to pick the equation for the unknown variable they are looking for
        select=int(input("\nSelect 1. for Time fo flight, 2. for maximum height, 3. for horizontal range: "))
        
        #if user wants to find the time of flight then solves for time of flight
        if select == 1:
            print("Enter the known variables:")
            self.initial_velocity = float(input("Enter the Initial velocity (m/s): "))
            self.theta = float(input("Enter theta (degrees): "))
            self.acceleration = float(input("Enter the Acceleration (m/s^2): "))
            self.flight_time_solve()
        #if user wants to find the maximum height then solves for maximum height
        elif select == 2:
            print("Enter the known variables:")
            self.initial_velocity = float(input("Enter the Initial velocity (m/s): "))
            self.theta = float(input("Enter theta (degrees): "))
            self.acceleration = float(input("Enter the Acceleration (m/s^2): "))
            self.max_height_solve()
        #if user wants to find the horizontal range then solves for horizontal range
        elif select == 3:
            print("Enter the known variables:")
            self.initial_velocity = float(input("Enter the Initial velocity (m/s): "))
            self.theta = float(input("Enter theta (degrees): "))
            self.acceleration = float(input("Enter the Acceleration (m/s^2): "))
            self.horizontal_range_solve() 
        #if the user input is invalid
        else:
            print("Invalid input.")
    
    #functions that calculate the unknown variable that the user is looking for
    def flight_time_solve(self):
        self.flight_time = (2*self.initial_velocity*math.sin(math.radians(self.theta)))/(self.acceleration)
        print("The flight time is: {:.3f} s".format(self.flight_time))
    def max_height_solve(self):
        self.max_height = (math.pow(self.initial_velocity, 2)*math.pow(math.sin(math.radians(self.theta)), 2))/(2*self.acceleration)
        print("The maximum height is: {:.3f} m".format(self.max_height))
    def horizontal_range_solve(self):
        self.horizontal_range = (math.pow(self.initial_velocity, 2)*math.sin(2*math.radians(self.theta)))/(self.acceleration)
        print("The horizontal range is: {:.3f} m".format(self.horizontal_range))
        
formula_choice=1 #initializing the formula choice
while formula_choice !=0: #while formula choice is True
    #prints the two options tat the user can pick from
    print("\n1. Kinematic Equations")
    print("2. Projectile Motion Equations")
    
    #prompts the user to pick which option they want
    formula_choice=int(input("Select the Equation Type. Select 0 to Exit: ")) 
    if formula_choice==1: #if user wants kinematics
        kinematics_solver = Kinematics()
        kinematics_solver.solve()
    elif formula_choice==2: #if user wants projectile motion
        projectile_motion_solver = Projectile_motion()
        projectile_motion_solver.solve()
    elif formula_choice==0: #if user wants to exit
        print("Exit.")
        break
    else: #if the user input is invalid
        print("Invalid input.")
```
